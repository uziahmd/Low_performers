```python
combined_data = pd.concat(DATA.values(), keys=DATA.keys(), names=['pcode', 'timestamp'])
```

```python
combined_data.index = combined_data.index.set_levels([combined_data.index.levels[0], pd.to_datetime(combined_data.index.levels[1])])
```

```python
def resample_user_data(df):
    return df.resample('15T').sum()

# Apply the resampling function to each user's data
resampled_data = combined_data.groupby(level='pcode').apply(resample_user_data)
```

```python
# Remove the extra level added by groupby
resampled_data.index = resampled_data.index.droplevel(0)
```

```python
def create_labels(df):
    labels = []
    for step_count in df:
        labels.append(step_count)
    return labels

# Apply labeling function to the resampled data
resampled_data['label'] = create_labels(resampled_data['steps'])
```
