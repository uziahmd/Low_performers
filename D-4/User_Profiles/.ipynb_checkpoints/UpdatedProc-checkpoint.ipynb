```python
import logging 
from tqdm import tqdm

def remove_duplicates(series: pd.Series, key: str) -> pd.Series:
    """
    Removes duplicate entries in a Series based on the index, keeping the first occurrence.

    Parameters:
    - series (pd.Series): The Series to process.
    - key (str): The key/name of the Series.

    Returns:
    - pd.Series: Series with duplicates removed.
    """
    initial_count = len(series)
    series_unique = series[~series.index.duplicated(keep='first')]
    duplicates_removed = initial_count - len(series_unique)
    if duplicates_removed > 0:
        logger.info(f"Removed {duplicates_removed} duplicate entries in {key}.")
    else:
        logger.info(f"No duplicates found in {key}.")
    return series_unique

def process_mobility_data(data: Dict[str, Any], key: str) -> pd.DataFrame:
    """
    Processes mobility-related data (LOC_LABEL, LOC_DST, LOC_CLS) and returns a cleaned DataFrame.

    Parameters:
    - data (Dict[str, Any]): The raw data dictionary.
    - key (str): The key to process ('LOC_LABEL', 'LOC_DST', 'LOC_CLS').

    Returns:
    - pd.DataFrame: Cleaned DataFrame for the specified key.
    """
    if key not in data:
        logger.error(f"Key '{key}' not found in data.")
        return pd.DataFrame()

    value = data[key]
    logger.info(f"Processing key: {key}, Type: {type(value)}")

    if not isinstance(value, pd.Series):
        logger.error(f"Expected pd.Series for '{key}', got {type(value)}. Skipping.")
        return pd.DataFrame()

    # Remove duplicates
    value_unique = remove_duplicates(value, key)

    # Convert Series to DataFrame
    df = value_unique.to_frame(name=key)

    # Handle missing values
    if key == 'LOC_LABEL':
        df[key].fillna('unknown', inplace=True)
    elif key == 'LOC_DST':
        df[key].fillna(0, inplace=True)  # Assume missing distance as 0
    elif key == 'LOC_CLS':
        df[key].fillna('UNKNOWN', inplace=True)

    # Ensure MultiIndex ['pcode', 'timestamp']
    if not isinstance(df.index, pd.MultiIndex):
        df.reset_index(inplace=True)
        df.set_index(['pcode', 'timestamp'], inplace=True)
        logger.info(f"Set MultiIndex to ['pcode', 'timestamp'] for {key}.")
    else:
        logger.info(f"DataFrame for {key} already has a MultiIndex.")

    # Convert 'timestamp' to datetime if not already
    if not pd.api.types.is_datetime64_any_dtype(df.index.get_level_values('timestamp')):
        df.index = df.index.set_levels(
            [
                df.index.get_level_values('pcode'),
                pd.to_datetime(df.index.get_level_values('timestamp'))
            ]
        )
        logger.info(f"Converted 'timestamp' to datetime for {key}.")

    # Remove duplicate indices
    df = df[~df.index.duplicated(keep='first')]
    logger.info(f"Removed duplicate index entries for {key}.")

    return df
```

```python
def process_additional_data(data: Dict[str, Any], key: str) -> pd.DataFrame:
    """
    Processes additional mobility data ('LOC_DST' or 'LOC_CLS') and returns a cleaned DataFrame.

    Parameters:
    - data (Dict[str, Any]): The raw data dictionary.
    - key (str): The key to process ('LOC_DST' or 'LOC_CLS').

    Returns:
    - pd.DataFrame: Cleaned DataFrame for the specified key.
    """
    return process_mobility_data(data, key)

def merge_dataframes(dfs: List[pd.DataFrame], keys: List[str]) -> pd.DataFrame:
    """
    Merges multiple DataFrames on their MultiIndex ['pcode', 'timestamp'].

    Parameters:
    - dfs (List[pd.DataFrame]): List of DataFrames to merge.
    - keys (List[str]): List of keys corresponding to the DataFrames.

    Returns:
    - pd.DataFrame: Merged DataFrame.
    """
    if not dfs:
        logger.error("No DataFrames provided for merging.")
        return pd.DataFrame()

    # Start with the first DataFrame
    merged_df = dfs[0]

    # Iteratively merge the remaining DataFrames
    for df, key in zip(dfs[1:], keys[1:]):
        merged_df = merged_df.join(df, how='outer')
        logger.info(f"Merged DataFrame with '{key}'.")

    # Handle any remaining missing values
    merged_df.fillna({
        'LOC_LABEL': 'unknown',
        'LOC_DST': 0,
        'LOC_CLS': 'UNKNOWN'
    }, inplace=True)

    return merged_df

def extract_mobility_features_enhanced(window_data: pd.DataFrame) -> Dict[str, Any]:
    """
    Extracts enhanced mobility-related features from a window of mobility data.

    Parameters:
    - window_data (pd.DataFrame): DataFrame containing 'LOC_LABEL', 'LOC_DST', and 'LOC_CLS' within the window.

    Returns:
    - Dict[str, Any]: Dictionary of extracted mobility features.
    """
    features = {}
    try:
        if window_data.empty:
            logger.debug("Window data is empty.")
            return features

        # Number of location changes
        loc_changes = window_data['LOC_LABEL'].ne(window_data['LOC_LABEL'].shift()).sum() - 1
        loc_changes = max(loc_changes, 0)
        features['LOC_CHANGE_CNT'] = loc_changes

        # Entropy of location labels
        counts = window_data['LOC_LABEL'].value_counts()
        entropy = sp.entropy(counts, base=2)
        features['LOC_ENTROPY'] = entropy

        # Time spent in each location
        window_data = window_data.sort_index()
        timestamps = window_data.index.get_level_values('timestamp')
        time_deltas = timestamps.to_series().diff().dt.total_seconds().fillna(0)

        loc_time = window_data.groupby('LOC_LABEL').apply(
            lambda x: time_deltas[x.index].sum()
        )

        for loc, dur in loc_time.items():
            features[f'LOC_TIME_{loc.upper()}'] = dur / 60  # Convert to minutes

        # Most frequent location
        if not counts.empty:
            mode_loc = counts.idxmax()
            features['LOC_MODE'] = mode_loc
        else:
            features['LOC_MODE'] = 'unknown'

        # Average and Std duration between location changes
        if loc_changes > 0:
            avg_dur = time_deltas.sum() / loc_changes
            features['LOC_AVG_DUR'] = avg_dur / 60  # minutes
            features['LOC_STD_DUR'] = time_deltas.std() / 60  # minutes
        else:
            features['LOC_AVG_DUR'] = 0
            features['LOC_STD_DUR'] = 0

        # Total Distance moved in window
        total_distance = window_data['LOC_DST'].sum()
        features['LOC_TOTAL_DST'] = total_distance  # Assuming distance is in meters

        # Average Distance per move
        num_moves = loc_changes
        if num_moves > 0:
            avg_distance = total_distance / num_moves
            features['LOC_AVG_DST_PER_MOVE'] = avg_distance
        else:
            features['LOC_AVG_DST_PER_MOVE'] = 0

        # Number of unique places visited
        unique_places = window_data['LOC_CLS'].nunique()
        features['LOC_NUM_PLCS_VIST'] = unique_places

    except Exception as e:
        logger.error(f"Error extracting enhanced mobility features: {e}")

    return features

def sliding_window_feature_extraction_mobility_enhanced(
    mobility_df: pd.DataFrame,
    window_size_minutes: int,
    step_size_minutes: int = None
) -> pd.DataFrame:
    """
    Extracts enhanced mobility features using a sliding window approach.

    Parameters:
    - mobility_df (pd.DataFrame): DataFrame with 'LOC_LABEL', 'LOC_DST', 'LOC_CLS' and MultiIndex ['pcode', 'timestamp'].
    - window_size_minutes (int): Size of the sliding window in minutes.
    - step_size_minutes (int, optional): Step size in minutes. Defaults to window_size_minutes.

    Returns:
    - pd.DataFrame: Extracted mobility features.
    """
    if step_size_minutes is None:
        step_size_minutes = window_size_minutes  # Default to non-overlapping windows

    extracted_features = []

    # Ensure the DataFrame is sorted by pcode and timestamp
    mobility_df = mobility_df.sort_index(level=['pcode', 'timestamp'])

    for pcode, group in tqdm(mobility_df.groupby(level='pcode'), desc="Processing Users"):
        # Reset 'pcode' level and keep 'timestamp' as a column
        group = group.reset_index(level='pcode', drop=True).reset_index()

        # Set 'timestamp' as datetime index
        group = group.set_index('timestamp')

        # Define the sliding window
        start_time = group.index.min()
        end_time = group.index.max()

        current_start = start_time
        current_end = current_start + pd.Timedelta(minutes=window_size_minutes)

        while current_end <= end_time:
            window = group[current_start:current_end]
            feature_row = {}
            feature_row['pcode'] = pcode
            feature_row['window_end_time'] = current_end

            mobility_feats = extract_mobility_features_enhanced(window)
            if mobility_feats:
                # Combine pcode, window_end_time with features
                combined_features = {'pcode': pcode, 'window_end_time': current_end}
                combined_features.update(mobility_feats)
                extracted_features.append(combined_features)

            # Move the window
            current_start += pd.Timedelta(minutes=step_size_minutes)
            current_end = current_start + pd.Timedelta(minutes=window_size_minutes)

    mobility_feature_df = pd.DataFrame(extracted_features)
    if not mobility_feature_df.empty:
        mobility_feature_df = mobility_feature_df.set_index(['pcode', 'window_end_time'])
    else:
        logger.error("No features extracted. Please check the data and feature extraction functions.")

    return mobility_feature_df
```

```python
def calculate_time_proportions(df: pd.DataFrame, loc_time_columns: List[str]) -> pd.DataFrame:
    """
    Calculates the proportion of time spent at each location.

    Parameters:
    - df (pd.DataFrame): DataFrame with location time columns.
    - loc_time_columns (List[str]): List of columns representing time spent at each location.

    Returns:
    - pd.DataFrame: DataFrame with proportion columns added.
    """
    df['TOTAL_TIME'] = df[loc_time_columns].sum(axis=1)
    for col in loc_time_columns:
        proportion_col = col.replace('LOC_TIME_', 'PROP_TIME_')
        df[proportion_col] = df[col] / df['TOTAL_TIME']
    return df
```

```python
# ---------------------------
# 14. Cumulative Distance Plot for All Selected Users on a Common Random Day
# ---------------------------

import random
import matplotlib.dates as mdates

def plot_cumulative_distance_common_day(features_df: pd.DataFrame, users: List[str]):
    """
    Selects a common random day across all selected users and plots the cumulative distance moved
    for each user on that day in the same graph.
    
    Parameters:
    - features_df (pd.DataFrame): DataFrame containing mobility features with MultiIndex ['pcode', 'window_end_time'].
    - users (List[str]): List of user codes to plot.
    """
    # Step 1: Extract available dates for each user
    user_dates = {}
    for user in users:
        user_data = features_df.reset_index()
        user_data_user = user_data[user_data['pcode'] == user]
        
        if user_data_user.empty:
            logger.warning(f"No data available for user {user}. Skipping.")
            continue
        
        # Extract unique dates
        user_data_user['date'] = user_data_user['window_end_time'].dt.date
        available_dates = set(user_data_user['date'].unique())
        user_dates[user] = available_dates
    
    # Step 2: Find the intersection of dates across all users
    common_dates = set.intersection(*user_dates.values()) if user_dates else set()
    
    if not common_dates:
        logger.error("No common dates found across all selected users. Cannot plot cumulative distances.")
        return
    
    # Step 3: Select a random common date
    random_common_date = random.choice(list(common_dates))
    print(f"Selected Common Date for All Users: {random_common_date}")
    
    plt.figure(figsize=(14, 8))
    
    # Step 4: Plot cumulative distance for each user on the selected common date
    for user in users:
        if user not in user_dates:
            continue  # Skip users with no data
        
        # Filter data for the user and the selected date
        user_data = features_df.reset_index()
        user_day_data = user_data[
            (user_data['pcode'] == user) & 
            (user_data['window_end_time'].dt.date == random_common_date)
        ].sort_values('window_end_time')
        
        if user_day_data.empty:
            logger.warning(f"No data available for user {user} on {random_common_date}. Skipping.")
            continue
        
        # Calculate cumulative distance
        user_day_data['Cumulative_Distance'] = user_day_data['LOC_TOTAL_DST'].cumsum()
        
        # Handle timezone consistency
        sample_timestamp = user_day_data['window_end_time'].iloc[0]
        if sample_timestamp.tz is None:
            # Timezone-naive
            random_date_ts = pd.Timestamp(random_common_date)
        else:
            # Timezone-aware
            user_timezone = sample_timestamp.tz
            random_date_ts = pd.Timestamp(random_common_date).tz_localize(user_timezone)
        
        # Create a time axis relative to the start of the day
        user_day_data['Time_of_Day'] = user_day_data['window_end_time'] - random_date_ts
        
        # Convert 'Time_of_Day' to total minutes for plotting
        user_day_data['Time_Minutes'] = user_day_data['Time_of_Day'].dt.total_seconds() / 60
        
        # Plot the cumulative distance
        sns.lineplot(
            x='Time_Minutes',
            y='Cumulative_Distance',
            data=user_day_data,
            marker='o',
            label=user
        )
    
    # Customize the plot
    plt.title(f'Cumulative Distance Moved by Selected Users on {random_common_date}', fontsize=16)
    plt.xlabel('Time of Day (Minutes since Start)', fontsize=14)
    plt.ylabel('Cumulative Distance (meters)', fontsize=14)
    plt.legend(title='User', fontsize=12)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Call the updated function to plot cumulative distance for all selected users on a common random day
plot_cumulative_distance_common_day(features_df_mobility_enhanced, selected_users)
```

```python



import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import itertools

# Assuming selected_users_df is already defined

# Step 1: Normalize each column using Min-Max normalization
normalized_df = selected_users_df.copy()
columns_to_normalize = 
['LOC_NUM_PLCS_VIST', 'LOC_TIME_NONE', 'LOC_TIME_WORK', 'LOC_TIME_EATING', 'LOC_TIME_SOCIAL', 'LOC_TIME_OTHERS', 'LOC_TIME_GYM']
for column in columns_to_normalize:
    min_val = selected_users_df[column].min()
    max_val = selected_users_df[column].max()
    normalized_df[column] = (selected_users_df[column] - min_val) / (max_val - min_val)

# Step 2: Group by 'pcode' and calculate mean and standard deviation for the normalized columns
grouped_mean = normalized_df.groupby('pcode').mean()[columns_to_normalize]
grouped_std = normalized_df.groupby('pcode').std()[columns_to_normalize]

# Step 3: Clip the lower bound of the error bars so they do not extend below zero
lower_bounds = grouped_mean - grouped_std
lower_bounds[lower_bounds < 0] = 0
adjusted_std = grouped_mean - lower_bounds

# Step 4: Define the plot size
fig, ax = plt.subplots(figsize=(14, 8))

# Step 5: Generate a large list of distinguishable colors
colors = itertools.cycle([
    'red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 
    'navy', 'magenta', 'gold', 'lime', 'teal'
])

# Step 6: Plot mean values with improved error bars for each column, using distinct colors
bar_width = 0.8 / len(columns_to_normalize)  # Adjust bar width to fit all columns
x = np.arange(len(grouped_mean.index))  # x positions for the groups

for idx, column in enumerate(columns_to_normalize):
    color = next(colors)  # Get the next color from the cycle
    # Plot the mean bars
    ax.bar(x + idx * bar_width, grouped_mean[column], width=bar_width, label=column, color=color, alpha=0.8)
    # Add error bars with a dotted line style
    ax.errorbar(x + idx * bar_width, grouped_mean[column], yerr=adjusted_std[column], fmt='none',
                ecolor='black', elinewidth=1, linestyle=':', capsize=2, alpha=0.5)


# Step 7: Set labels, title, and legend (legend positioned outside the figure)
ax.set_xlabel("Pcode")
ax.set_ylabel("Normalized Values")
ax.set_title("Mean/Std of Normalized Mobility Features Across Users (normalized by feature)")

# Configure y-axis
ax.set_ylim(0, 0.4)  # Set y-axis range from 0 to 1 for normalized values


# Set xticks at the center of grouped bars
ax.set_xticks(x + bar_width * (len(columns_to_normalize) / 2))
ax.set_xticklabels(grouped_mean.index, rotation=45)

# Place legend outside the plot
ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1), borderaxespad=0.)

# Step 8: Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import itertools
import matplotlib.pyplot as plt

# Assuming selected_users_df is already defined and 'pcode' is the user identifier

# Define the columns to be used for calculating Cohen's D
selected_columns = [
    'LOC_NUM_PLCS_VIST', 'LOC_TIME_NONE', 'LOC_TIME_WORK',
    'LOC_TIME_EATING', 'LOC_TIME_SOCIAL', 'LOC_TIME_OTHERS', 'LOC_TIME_GYM'
]

# Step 1: Create a function to calculate absolute Cohen's D
def cohen_d(x1, x2):
    x1 = x1.dropna()
    x2 = x2.dropna()
    n1, n2 = len(x1), len(x2)
    if n1 < 2 or n2 < 2:
        return np.nan
    mean1, mean2 = x1.mean(), x2.mean()
    var1, var2 = x1.var(ddof=1), x2.var(ddof=1)
    pooled_std = np.sqrt(
        ((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2)
    )
    if pooled_std == 0:
        return np.nan
    else:
        return abs((mean1 - mean2) / pooled_std)

# Step 2: Extract unique user codes and initialize a matrix for mean absolute Cohen's D values
user_codes = selected_users_df['pcode'].unique()
cohen_d_matrix = pd.DataFrame(index=user_codes, columns=user_codes)

# Step 3: Calculate mean absolute Cohen's D for each pair of users across selected columns
for user1, user2 in itertools.product(user_codes, repeat=2):
    if user1 == user2:
        cohen_d_matrix.loc[user1, user2] = 0  # Diagonal set to zero
        continue
    data1 = selected_users_df[selected_users_df['pcode'] == user1][selected_columns]
    data2 = selected_users_df[selected_users_df['pcode'] == user2][selected_columns]
    
    d_values = []
    for column in selected_columns:
        d = cohen_d(data1[column], data2[column])
        if not np.isnan(d):
            d_values.append(d)
    # Compute mean absolute Cohen's D
    if len(d_values) > 0:
        mean_d = np.mean(d_values)
    else:
        mean_d = np.nan
    cohen_d_matrix.loc[user1, user2] = mean_d

# Convert to numeric
cohen_d_matrix = cohen_d_matrix.apply(pd.to_numeric, errors='coerce')

# Replace infinite values
cohen_d_matrix.replace([np.inf, -np.inf], np.nan, inplace=True)

# Ensure all values are non-negative (since we took absolute values)
cohen_d_matrix = cohen_d_matrix.abs()

# Step 4: Set the diagonal values explicitly to 0
np.fill_diagonal(cohen_d_matrix.values, 0)

# Step 5: Create a heatmap using Matplotlib's `pcolormesh()`
data = cohen_d_matrix.values
labels = cohen_d_matrix.index

fig, ax = plt.subplots(figsize=(16, 12))

# Define the color map and draw the heatmap using `pcolormesh`
vmin, vmax = 0, 1  # Set fixed range for Cohen's D thresholds (0.2, 0.5, 0.8 as intrinsic thresholds)
cmap = plt.get_cmap('RdYlBu_r')  # Vibrant color map for better contrast
heatmap = ax.pcolormesh(data, cmap=cmap, edgecolors='k', linewidth=0.5, vmin=vmin, vmax=vmax)

# Add color bar
cbar = plt.colorbar(heatmap)
cbar.set_label("Mean Absolute Cohen's D", fontsize=14)

# Set axis labels and ticks
ax.set_xticks(np.arange(data.shape[1]) + 0.5)
ax.set_yticks(np.arange(data.shape[0]) + 0.5)
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=10)
ax.set_yticklabels(labels, fontsize=10)

# Annotate each cell with the value from `cohen_d_matrix`
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        ax.text(j + 0.5, i + 0.5, f'{data[i, j]:.2f}', 
                ha='center', va='center', color='black', fontsize=12)

# Set title and layout
plt.title("Mean Absolute Cohen's D Between Users' Mobility Features", fontsize=16)
plt.tight_layout()
plt.show()

```
