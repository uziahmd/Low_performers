```python
from Funcs.Utility import *
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = os.path.join(PATH_INTERMEDIATE, 'calorie_count_binary_personal-15min.pkl')
X, y, groups, t, datetimes = load(p)
```

```python
#Divide the features into different categories
feat_current = X.loc[:,[('#VAL' in str(x)) or ('ESM#LastLabel' in str(x)) for x in X.keys()]]  
feat_dsc = X.loc[:,[('#DSC' in str(x))  for x in X.keys()]]  
feat_yesterday = X.loc[:,[('Yesterday' in str(x))  for x in X.keys()]]  
feat_today = X.loc[:,[('Today' in str(x))  for x in X.keys()]]  
feat_sleep = X.loc[:,[('Sleep' in str(x))  for x in X.keys()]]  
feat_time = X.loc[:,[('Time' in str(x))  for x in X.keys()]]  
feat_pif = X.loc[:,[('PIF' in str(x))  for x in X.keys()]]  
feat_ImmediatePast = X.loc[:,[('ImmediatePast_15' in str(x))  for x in X.keys()]]
#Divide the time window features into sensor/past stress label
feat_current_sensor = X.loc[:,[('#VAL' in str(x))  for x in X.keys()]]  
feat_current_ESM = X.loc[:,[('ESM#LastLabel' in str(x)) for x in X.keys()]]  
feat_ImmediatePast_sensor = feat_ImmediatePast.loc[:,[('ESM' not in str(x)) for x in feat_ImmediatePast.keys()]]  
feat_ImmediatePast_ESM = feat_ImmediatePast.loc[:,[('ESM'  in str(x)) for x in feat_ImmediatePast.keys()]]  
feat_today_sensor = feat_today.loc[:,[('ESM' not in str(x))  for x in feat_today.keys()]]  
feat_today_ESM = feat_today.loc[:,[('ESM'  in str(x)) for x in feat_today.keys()]]  
feat_yesterday_sensor = feat_yesterday.loc[:,[('ESM' not in str(x)) for x in feat_yesterday.keys()]]  
feat_yesterday_ESM = feat_yesterday.loc[:,[('ESM'  in str(x)) for x in feat_yesterday.keys()]]
```

```python
feat_baseline = pd.concat([ feat_time,feat_dsc,feat_current_sensor, feat_ImmediatePast_sensor],axis=1)
feat_final = pd.concat([feat_baseline  ], axis=1)
# # Fill NaN values with zeros
feat_final = feat_final.fillna(0)
X = feat_final
cats = X.columns[X.dtypes == bool]
```

```python
X = X[[col for col in X.columns if not col.endswith('#ImmediatePast_15')]]
cn = X.columns.tolist()
print(cn)
```

```python
X['SCR_EVENT#DSC']
```

```python
df = pd.DataFrame(X)
df['UserId'] = groups
df['Timestamp'] = pd.to_datetime(datetimes)
```

```python
df
```

```python
# Select 7 random unique User IDs
random_users = np.random.choice(df['UserId'].unique(),3 , replace=False)
```

```python
random_users
```

```python
# Filter DataFrame for these users
filtered_df = df[df['UserId'].isin(random_users)]

# Extract the hour and minute as a combined decimal time representation for each timestamp
filtered_df['HourOfDay'] = filtered_df['Timestamp'].dt.hour + filtered_df['Timestamp'].dt.minute / 60.0

# Aggregate battery level by averaging over each time point for each user
aggregated_df = filtered_df.groupby(['UserId', 'HourOfDay'])['BAT_TMP#VAL'].mean().reset_index()

# Plot battery level trends for each user separately
plt.figure(figsize=(12, 8))

for user in random_users:
    user_data = aggregated_df[aggregated_df['UserId'] == user]
    plt.plot(user_data['HourOfDay'], user_data['BAT_TMP#VAL'], marker='o', linestyle='-', label=f'User {user}')

plt.xlabel('Time of Day (Hour)')
plt.ylabel('Average Battery Level (BAT_TMP#VAL)')
plt.title('Average Battery Level Fluctuation Throughout the Day for 3 Random Users (Aggregated Across Days)')
plt.legend(title='UserId')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter DataFrame for these users and for non-null 'PWR#DSC=ACTIVATE' (indicating power save activation)
filtered_df = df[df['UserId'].isin(random_users)]
power_save_df = filtered_df[filtered_df['PWR#DSC=ACTIVATE'].notna()]

# Extract hour of day to analyze power save activation times throughout the day
power_save_df['HourOfDay'] = power_save_df['Timestamp'].dt.hour + power_save_df['Timestamp'].dt.minute / 60.0

# Count the number of times power save mode was activated for each user by hour
activation_counts = power_save_df.groupby(['UserId', 'HourOfDay']).size().reset_index(name='ActivationCount')

# Plot the power save activation count for each user
plt.figure(figsize=(12, 8))

for user in random_users:
    user_data = activation_counts[activation_counts['UserId'] == user]
    plt.plot(user_data['HourOfDay'], user_data['ActivationCount'], marker='o', linestyle='-', label=f'User {user}')

plt.xlabel('Time of Day (Hour)')
plt.ylabel('Power Save Activation Count (PWR#DSC=ACTIVATE)')
plt.title('Power Save Activation Times Throughout the Day for 3 Random Users')
plt.legend(title='UserId')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
